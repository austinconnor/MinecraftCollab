```python
import os
import pandas as pd
import numpy as np
```

```python
def list_to_dic(lst):
    index = [i for i in range(len(lst))]
    zipped = list(zip(lst,index))
    ret = {zipped[i][0]: zipped[i][1] for i in range(len(zipped))}
    return ret
```

```python
def process_data(world, event):
    event_file = '../../data/' + world + '/' + event + '.csv'
    event_df = pd.read_csv(event_file)
    display(event_df)
    
    # drop unnecessary columns
    event_df.drop(columns=['_id','event'], inplace=True)
    display(event_df)
    
    unique_players = list(event_df['player'].unique())
    print(unique_players)
    
    player_mapping = list_to_dic(unique_players)
    
    event_df['player'] = event_df['player'].map(player_mapping)
    
    unique_players = list(event_df['player'].unique())
    print(unique_players)
    
    event_df.rename(columns={'location.x': 'x', 'location.y': 'y', 'location.z': 'z'}, inplace=True)
    
    return (event_df, unique_players)
```

```python
def make_csv(move, players, world):
    for player in players:
        df = move[move['player'] == player]
        print('BEFORE: ' + str(len(df)))
        #drop all odd numbered rows
    #     drop_i = [i for i in list(range(len(df))) if i % 2 == 1]
    #     df.drop(drop_i, inplace=True)
        df = df[df.index % 20 == 0]
        print('AFTER: ', str(len(df)))

        filepath = '../../public/data/' + world + "/"
        filename = filepath + str(player) + '_move.csv'
        df.to_csv(filename, index=False)
```

## Movement Overworld

```python
(df, unique_players) = process_data('overworld', 'PlayerMoveEvent')
```

```python
make_csv(df, unique_players, 'overworld')
```

## Movement Nether

```python
(df, unique_players) = process_data('nether', 'PlayerMoveEvent')
```

```python
make_csv(df, unique_players, 'nether')
```

## Movement End

```python
(df, unique_players) = process_data('end', 'PlayerMoveEvent')
```

```python
make_csv(df, unique_players, 'end')
```
